```python
import cv2
import google.colab.patches as colab
import numpy
```

```python
in_image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/license_plate.bmp", 0)
colab.cv2_imshow(in_image)
```

```python
grad_x = cv2.Sobel(in_image, cv2.CV_16S, 1, 0)
grad_y = cv2.Sobel(in_image, cv2.CV_16S, 0, 1)
```

compute $|G_x|$ and $|G_y|$

```python
abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv2.convertScaleAbs(grad_y)
```

compute $1.0\times |G_x|$ + $1.0 \ \times|G_y|$

```python
grad_image = cv2.addWeighted(abs_grad_x, 1.0, abs_grad_y, 1.0, 0)
```

```python
otsu_th, edge_image = cv2.threshold(grad_image, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
colab.cv2_imshow(edge_image)
```
